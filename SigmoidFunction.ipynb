```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.linspace(-10, 10, 100)
b = 1/(1 + np.exp(-a))
plt.style.use("seaborn-v0_8-dark")
plt.plot(a,b,lw=4,marker="*",color="purple",markersize=10,markerfacecolor="red")
plt.xlabel("a",fontsize=12,color="blue")
plt.ylabel("Sigmoid(A)",fontsize=12,color="blue")
plt.title("Sigmoid Function\n",fontsize=15,color="blue")
plt.show()
```

```python

def sig(x):
 return 1/(1 + np.exp(-x))
```

```python
y = 1.0
print("Value is %.2f"%sig(y))
```

```python
m = -1.0
print("Value is %.2f"%sig(m))
```

```python
n = -0.00478
print("Value is %.2f"%sig(n))
```
