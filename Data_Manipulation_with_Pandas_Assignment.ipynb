# Design Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr,cmap="hsv")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, vmin=-1, vmax=1, annot=True,cmap="inferno")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
mask = np.tril(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, annot=True, cmap='summer')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, annot=True, cmap='turbo')
plt.show()
```
